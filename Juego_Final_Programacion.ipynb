Juego Picas y Fijas.

```python
import random

secreto = ""
while len(secreto) < 4:
  d = str(random.randint(0,9))
  if len(secreto) == 0 and d == "0":
      continue
  if d not in secreto:
      secreto += d

print("picas y fijas")
print("Adivina mi numero de 4 digitos")

while True:
    intento = input("numero: ").strip()

    if len(intento) != 4 or not intento.isdigit():
        print("dame 4 digitos")
        continue
    picas = 0
    fijas = 0

    for i in range(4):
        if intento[i] == secreto[i]:
           fijas += 1
        elif intento[i] in secreto:
           picas += 1

    print("picas:", picas,"fijas:", fijas)

    if fijas == 4:
       print("muy bien adivinaste mi numero:", secreto)
       break
```
